# Understanding Images and Image Processing with OpenCv:

<b>Import required libraries<b>

```python
import numpy as np
import pandas as np
import matplotlib
import matplotlib.pyplot as plt

%matplotlib inline
```

<b>Import image loading and processing libraries<b>

```python
from PIL import Image
import cv2
```

<b>Reading Image using PIL<b>

```python
img = Image.open('./Data/birds.jpg')
img
```

<b>Loading image using matplotlib<b>

```python
img_mat = matplotlib.image.imread('./Data/birds.jpg')
img_mat
```

<b>Loading image using openCv<b>

```python
img_cv = cv2.imread('./Data/birds.jpg')
img_cv
```

<b>Description of Image<b>

```python
img_cv.shape
```

```python
img_mat.shape
```

<b>OpenCv by default renders images in BGR format as shown below:<b>

```python
plt.figure(figsize=(12,5))
plt.imshow(img_cv)
```

<b>Matplotlib array uses RGB format for rendering images<b>

```python
plt.figure(figsize=(12,5))
plt.imshow(img_mat)
```

<b>Analyze image using matplotlib:<b>

```python
img_mat.shape
```

<b>Split array into 3 parts, each representing a single primary color<b>

```python
red = img_mat[:,:,0]
green = img_mat[:,:,1]
blue = img_mat[:,:,2]
```

```python
red.shape, green.shape, blue.shape
```

<b>Display individual color arrays:<b>

```python
plt.figure(figsize=(16,5))
plt.subplot(1,3,1)
plt.title('Red Color Region')
plt.imshow(red)
plt.subplot(1,3,2)
plt.title('Green Color Region')
plt.imshow(green)
plt.subplot(1,3,3)
plt.title('Blue Color Region')
plt.imshow(blue)
```

<b>Rendering individual color arrays in greyscale:<b>

```python
plt.figure(figsize=(16,5))
plt.subplot(1,3,1)
plt.title('Red Color Region')
plt.imshow(red,cmap='gray')
plt.subplot(1,3,2)
plt.title('Green Color Region')
plt.imshow(green,cmap='gray')
plt.subplot(1,3,3)
plt.title('Blue Color Region')
plt.imshow(blue,cmap='gray')
```

<b>Converting 3D color image to 2D grayscale image using openCv:<b>

```python
mat_grayscale = cv2.cvtColor(img_mat,cv2.COLOR_RGB2GRAY)
cv_grayscale = cv2.cvtColor(img_cv,cv2.COLOR_BGR2GRAY)
```

<b>Visualizing the grayscale images<b>

```python
plt.figure(figsize=(16,4))

plt.subplot(1,2,1)
plt.title('MatPlotLib Rendered Image')
plt.imshow(mat_grayscale)

plt.subplot(1,2,2)
plt.title('OpenCv Rendered Image')
plt.imshow(cv_grayscale)
```

```python
plt.figure(figsize=(16,4))

plt.subplot(1,2,1)
plt.title('MatPlotLib Rendered Image')
plt.imshow(mat_grayscale,cmap='gray')

plt.subplot(1,2,2)
plt.title('OpenCv Rendered Image')
plt.imshow(cv_grayscale,cmap='gray')
```
