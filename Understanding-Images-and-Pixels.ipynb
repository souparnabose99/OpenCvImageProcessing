```python
plt.figure(figsize=(8,8))
plt.subplot(1,2,1)
plt.title('Array as Colour Image')
plt.imshow(arr2)
plt.subplot(1,2,2)
plt.title('Array as Grayscale Image')
plt.imshow(arr2,cmap='gray')
```

Images have pixels.
Each pixel has a range of 0-(2^n -1).
eg- for 8 bit image, it is 0-255.
Lower the value, darker the color.
Higher the value, lighter the color.
eg- for grayscale black is 0 and white is 255.
So, images are basically an array of numbers.

<b>Read an image and visualize it<b>

```python
img_cv = cv2.imread('./Data/birds.jpg')

plt.figure(figsize=(16,4))
plt.subplot(1,2,1)
plt.title('Colour Image')
plt.imshow(img_cv)
plt.subplot(1,2,2)
plt.title('Grayscale Image')
plt.imshow(img_cv,cmap='gray')
#Since image is not converted to grayscale yet, cmap='gray' will show the same results
```

```python
img_cv.shape
```

<b>Convert color image to grayscale<b>

```python
gray_img_cv =  cv2.cvtColor(img_cv,cv2.COLOR_BGR2GRAY)

plt.figure(figsize=(16,4))
plt.subplot(1,2,1)
plt.title('Colour Image')
plt.imshow(gray_img_cv)
plt.subplot(1,2,2)
plt.title('Grayscale Image')
plt.imshow(gray_img_cv,cmap='gray')
```

```python
gray_img_cv.shape
```

<b>Slice different parts of the image and analyze<b>

```python
sliced_gray = gray_img_cv[0:10,0:10]

plt.figure(figsize=(8,8))
plt.subplot(1,2,1)
plt.title('Colour Image of slice')
plt.imshow(sliced_gray)
plt.subplot(1,2,2)
plt.title('Grayscale Image of slice')
plt.imshow(sliced_gray,cmap='gray')
```

```python
sliced_gray2 = gray_img_cv[50:350,300:600]

plt.figure(figsize=(8,8))
plt.subplot(1,2,1)
plt.title('Colour Image of slice')
plt.imshow(sliced_gray2)
plt.subplot(1,2,2)
plt.title('Grayscale Image of slice')
plt.imshow(sliced_gray2,cmap='gray')
```

## Resizing Images:

```python
plt.imshow(img_cv)
```

```python
img_cv.shape
```

Resizing are of 2 types, shrinking and enlarging.

<b>Shrink image:<b>

```python
shrunk_img_cv = cv2.resize(img_cv,(350,500),cv2.INTER_AREA)
shrunk_img_cv.shape
```

```python
plt.figure(figsize=(10,6))
plt.imshow(shrunk_img_cv)
```

###### <b>Enlarge Image:<b>

```python
enlarge_img_cv = cv2.resize(img_cv,(1200,2000),cv2.INTER_CUBIC)
enlarge_img_cv.shape
#For enlarging images from small size, INTER_CUBIC is better than INTER_AREA
```

```python
plt.figure(figsize=(10,6))
plt.imshow(enlarge_img_cv)
```

<b>Enlarging image results in insertion of more no of duplicate values, hence image might become blurred<b>
